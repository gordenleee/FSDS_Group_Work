```python
# 设置图表的整体风格为ggplot
plt.style.use('ggplot')

# 准备数据：单独的年份数据
data_by_year = {
    2019: data[data['year'] == 2019]['price'].dropna(),
    2020: data[data['year'] == 2020]['price'].dropna(),
    2021: data[data['year'] == 2021]['price'].dropna(),
    2022: data[data['year'] == 2022]['price'].dropna()
}

# 设置直方图的间隔数量
bins = 200

# 创建直方图
fig, axs = plt.subplots(2, 2, figsize=(15, 12), constrained_layout=True)

# 为每一个年份创建一个直方图
for i, (year, prices) in enumerate(data_by_year.items()):
    ax = axs[i//2, i%2]
    ax.hist(prices, bins=bins, color='navy', edgecolor='white')
    ax.set_title(f'{year} price')
    ax.set_xlabel('price')
    ax.set_ylabel('numbers')
    ax.set_ylim([0, 50000])
    ax.set_xlim([0, 2000])
    ax.grid(True)

# 显示图表
plt.show()
```

```python
# 设置图表的整体风格为ggplot
plt.style.use('ggplot')

# 准备数据：单独的年份数据，这里假设有一个 'price' 列和 'year' 列
price_data = [data[data['year'] == y]['price'].dropna() for y in range(2019, 2024)]

# 创建箱形图并处理离群点
plt.figure(figsize=(12, 8))

# 为了处理离群点，可以使用参数 sym='' 来忽略它们
plt.boxplot(price_data, patch_artist=True, sym='')

# 设置图表标题和轴标签
plt.title(' (2019-2023)box plot')
plt.xlabel('year')
plt.ylabel('price')

# 设置x轴刻度标签
plt.xticks([1, 2, 3, 4, 5], ['2019', '2020', '2021', '2022', '2023'])

# 显示图表
plt.show()
```

```python
# 创建箱形图并处理离群点，同时添加趋势线

# 设置图表风格
plt.style.use('ggplot')

# 计算每年价格的中位数，用于绘制趋势线
median_prices = [year_data.median() for year_data in price_data]

# 创建箱形图
plt.figure(figsize=(10, 8))
boxplots = plt.boxplot(price_data, patch_artist=True, sym='', labels=['2019', '2020', '2021', '2022', '2023'])

# 绘制趋势线
plt.plot(range(1, 6), median_prices, marker='o', color='red', label='Median Price Trend')

# 设置图表标题和轴标签
plt.title('Annual Price Distribution with Trend Line (2019-2023)')
plt.xlabel('Year')
plt.ylabel('Price')

# 显示图例
plt.legend()

# 显示图表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 准备数据分析
# 房源数量和价格波动分析
annual_listing_counts = data.groupby('year').size()
annual_price_means = data.groupby('year')['price'].mean()

# 房源类型市场份额分析
room_type_annual_distribution = data.groupby(['year', 'room_type']).size().unstack(fill_value=0)
room_type_annual_shares = room_type_annual_distribution.div(room_type_annual_distribution.sum(axis=1), axis=0)

# 设置图表风格
sns.set(style="whitegrid")

# 创建图表布局
fig, axes = plt.subplots(2, 2, figsize=(18, 18), constrained_layout=True)

# 房源数量和价格波动分析图表
# 折线图 - 房源数量和平均价格变化趋势
sns.lineplot(data=annual_listing_counts, ax=axes[0, 0], marker='o')
sns.lineplot(data=annual_price_means, ax=axes[0, 1], color='orange', marker='o')

axes[0, 0].set_title('Annual Number of Listings')
axes[0, 1].set_title('Annual Average Price')

# 房源类型市场份额分析图表
# 堆叠柱状图 - 不同房源类型的市场份额变化
room_type_annual_shares.plot(kind='bar', stacked=True, ax=axes[1, 0], colormap='viridis')



# 显示图表
plt.show()
```

```python
# 设置图表风格
sns.set(style="whitegrid")

# 创建每个年份的饼图，分别展示不同房源类型的市场份额
fig, axs = plt.subplots(3, 2, figsize=(14, 21))  # 3行2列的子图布局

# 逐年绘制饼图
for i, year in enumerate(range(2019, 2024)):
    # 选择子图位置
    ax = axs[i//2, i%2]
    # 提取当前年份的数据
    year_data = room_type_annual_distribution.loc[year]
    # 绘制饼图
    year_data.plot(kind='pie', autopct='%1.1f%%', startangle=90, ax=ax, colors=sns.color_palette('pastel'))
    # 设置标题
    ax.set_title(f'{year} Room Type Market Share')
    # 隐藏y轴标签
    ax.set_ylabel('')

# 删除多余的子图位（如果有的话）
if len(axs.flatten()) > len(range(2019, 2024)):
    for i in range(len(range(2019, 2024)), len(axs.flatten())):
        fig.delaxes(axs.flatten()[i])

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```

```python
#处理异常数据

# 删除价格为0的记录
data = data[data['price'] > 0]

# 定义价格的2%和98%的分位数
lower_percentile = data['price'].quantile(0.01)
upper_percentile = data['price'].quantile(0.99)

# 移除价格处于前2%和后2%的数据
data_filtered = data[(data['price'] > lower_percentile) & (data['price'] < upper_percentile)]
```
